### Find the number of movies released between 1950 and 1960
##### first we need understand how brackets and conditional operator works in case of pandas. here in this question we have check two expression/condition using logical operator. 
##### now we need to understand when we are comparing two or more expression/condition using any logical operator we have to specify the expression/condition in round brackets as i have done below. if in case you just have only one expression/condition to check then you dont have to use round bracket  
##### now talking about code we have to specify two expression her as (data['Year']>=1950) and (data['Year']<=1960), here 'year' is a series of movies data set which indicate the Year of movie release. 

```python
print("number of movies released between 1950 and 1960 :")

#len () returns the length (the number of items) of an object.
len(data[(data['Year']>=1950) & (data['Year']<=1960)])
```

##### if you want name of the movies just not the numbers that are released between 1950 and 1960 then you can just remove the len() then you will see the whole data frame of the movies that are release between 1950 and 1960 , but in case you dont want the whole data set and just name of movies then you have to save that result in temporary data frame or else you can over-right real data frame but doin that you will loose the real data that you have loaded before.

```python
#data1 is temprory data frame it can be over-right later
data1 = data[(data['Year']>=1950) & (data['Year']<=1960)]

data1['name']
```

### Handling null values in data set.

```python
#isnull () function detect missing values in the given series object

data.isnull()
```

```python
#here isnull funtion will fine the null values and sum function will sum them and give us acurate number of missing value

data.isnull().sum()
```

```python
#dropna( ) will drop the null vlaues.

data.dropna()
```

##### dropna( ) will drop the null values in the data set. 
##### this function takes parameter as i have given below
##### dropna(how='any') means drop the value of row if any of the value is null

```python
data.dropna(how='any')
```

##### dropna(how='all') means drop the value of row if all of the value is null.( if all values of row is not null then it will not drop that values )

```python
data.dropna(how='all')
```

##### here we are pasing the list to the function saying that drop the null values if rating or duration values are null 
##### here we are using 'any' to the how argument thats why it will drop if any of the value is null

```python
data.dropna(subset=['Rating','duration'], how = 'any')
```

##### here we are pasing the list to the function saying that drop the null values if rating and duration values are null 
##### here we are using 'all' to the how argument thats why it will only drop the values if both the values are null

```python
data.dropna(subset=['Rating','duration'], how = 'all')
```

### filling the null values 
#### we fill null values using fillna() function 
#### here we are replacing the null value from mean 

```python
# mean() is use to find the mean 

M_value=int(data['Rating'].mean())
```

```python
M_value
```

```python
# now we are replacing the zero with mean value

data['Rating'].fillna( M_value , inplace=True)
```
